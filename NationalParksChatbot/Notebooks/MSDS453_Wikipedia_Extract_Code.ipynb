```python
import requests
from bs4 import BeautifulSoup
resp = requests.get('https://en.wikipedia.org/wiki/List_of_national_parks_of_the_United_States')
```

```python
text = resp.text
soup = BeautifulSoup(text)
parks = []
park_rows_th = soup.select('th>a[title]')
for park in park_rows_th:
  parks.append(park['title'])
park_rows_td = soup.select('td[scope="row"]>a[title]')
for park in park_rows_td:
  parks.append(park['title'])
parks = sorted(parks)
parks
```

```python
import wikipedia
import re
import pandas as pd

def get_wiki_text(park):
  try:
    wiki = wikipedia.page(park)
    text = wiki.content
    title = wiki.title
  except:
    try:
      wiki = wikipedia.page(park+ ' - Wikipedia')
      text = wiki.content
      title = wiki.title
    except:
      return park, None
  return title, text

def clean_wiki_text(text):
  text = re.sub(r'==.*?==+', '', text)
  text = text.replace('\n', '')
  return text

def get_parks_dict(park_titles):
  parks_dict = {}
  for park in park_titles:
    title, text = get_wiki_text(park)
    if text is None:
      parks_dict[title] = None
      continue
    text = clean_wiki_text(text)
    parks_dict[title] = text
  return parks_dict

def get_parks_df(parks_dict):
  parks_df = pd.DataFrame(parks_dict.items(), columns=['park', 'text'])
  return parks_df

parks_dict = get_parks_dict(parks)
parks_df = get_parks_df(parks_dict)
```
